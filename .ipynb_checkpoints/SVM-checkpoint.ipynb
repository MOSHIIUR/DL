# Finding SVM Loss 

```python
# calculate svm loss using naive approach
def naive_svm_loss(weight, train_img, train_labels, reg):
    loss = 0.0
    dW = np.zeros(weight.shape)
    
    num_train = train_img.shape[0]
    num_classes = weight.shape[1]
    for i in range(num_train):
    #     finding an(ith) img score using dot product
        pred_score = train_img[i].dot(weight)
#         print(pred_score)

    #     find ith img correct label from the train_labels
        idx_correct_class = train_labels[i]
        corrected_class_score = pred_score[idx_correct_class]
#         print(f'Corrected Class index: {idx_correct_class} & score: {corrected_class_score}')

        for j in range(num_classes):

            if j == idx_correct_class: #skipping the right class from calculating loss
                continue 

            loss_mergin = pred_score[j] - corrected_class_score + 1 #calculate the loss

            if loss_mergin > 0:
                loss += loss_mergin
                dW[:, j] += train_img[i] #append the gradient of wrong class
                dW[:, idx_correct_class] -= train_img[i] #append gradient of the wrong class        
        

#         print(f'Loss: {loss}')
#         print('-----------------------------------')
    
    loss /= num_train #get the average loss
    penalty = reg*np.sum(weight**2) #calculating regularization
    loss +- penalty #add penalty to the average loss
    
    dW /= num_train #scale/normalize the gradient
    dW += 2*reg*weight #append penalty to each of the gradient
    
    return loss, dW

    
```

# Bug in SVM loss

```python
# lets more generalize for linear classifer y = w(x)
# let
x = np.array((1, 1,1 ,1))
w1 = np.array((1, 0, 0, 0))
w2 = np.array((0.25, 0.25, 0.25, 0.25))



print(f'Scores: {x.dot(w1)}')
print(f'Scores: {x.dot(w2)}')

# regularization = weight-decay
# regularization stength
reg_strength = 1
print(f'Regularization(Penalty) for w1: {reg_strength*np.sum(w1*w1)}')
print(f'Regularization(Penalty) for w2: {reg_strength*np.sum(w2*w2)}')

```

As you can see above for both `w1` and `w2` we get the same `score`. As they both get the same score, the `average_loss` for both of the weight would be simillar as well. But as we add some penalty known as `Regularization` for w1 this penalty is higher than the w2. As if we are giving the preferences to the weight w1.

**But here this question needs to be asked why do we care? wasn't our objective was to find a weight that would give us a good prediction? Now we not only need a weight to give us good predicted score but also that weight needs to be bit nice as well? most iportant question among them is why defused one(w2) but not w1 get the references?**

So regularization and weighst are like too people who can not tolearte each other. Always invloves with conflict. Weight is like I am here to just fit the data, but regularization does not approve that, he wants the weight to be nice as well. And to do this they are fighting with each other. **But still this does not answer our questions!!!!!!** 🤔

Lets start with the last question `why the preference over w2 rather than w1?`
- `The most appealing property is that penalizing large weights tends to improve generalization, because it means that no input dimension can have a very large influence on the scores all by itself` __[cs231n](https://cs231n.github.io/linear-classify/)__




Ultimately that said averything there is to why adding this penalty!!! So, the effects after applying the penalty the scores/accuracy might be less, or other way to say it might not correctly find all the data in training time but the test case performances gets better. So, basically not only we want to fit the weight to training data, we as well want it to be fit with the testing data. That is when bias and variance concept comes along!!!__[StateQuest-Bias&Variance](https://www.youtube.com/watch?v=EuBBz3bI-aA)__ I would like to have my model both low in `Bias` and `variance`. For that I need a simple w. A weight(`w1`-was dominated on one particular pixels) that is bias towards a particular training data and a weight(`w2`:it was defused along the axis) that is not my regularization would prefer the one which is not bias towards. Hence, this would reduce some accuracy but would performs/generalize better.

An additional question `why not also regularize the bias then??`

```python
import numpy as np

# Define data
np.random.seed(42)  # For reproducibility
X = np.random.randn(3, 4)  # 3 examples, 4 features
W = np.random.randn(4, 3)  # Weights
y = np.array([0, 1, 2])  # Correct labels
reg = 0.1  # Regularization strength

# Calculate scores
scores = X.dot(W)

# Compute correct class scores
correct_class_scores = scores[np.arange(3), y]

# Compute margins
margins = np.maximum(0, scores - correct_class_scores[:, np.newaxis] + 1)
margins[np.arange(3), y] = 0  # Set margins of the correct class to zero

# Initialize binary margins
binary_margins = margins
binary_margins[margins > 0] = 1

# Calculate row sum
row_sum = np.sum(binary_margins, axis=1)

# Set margins of the correct class to -row_sum
binary_margins[np.arange(3), y] = -row_sum

# Calculate gradient
dW = (X.T).dot(binary_margins)
dW /= 3  # Divide by the number of examples
dW += reg * W  # Regularization term

# Display results
print("Scores:")
print(scores)
print("\nMargins:")
print(margins)
print("\nRow Sum:")
print(row_sum)
print("\nTrain:")
print(X.T)
print("\nBinary Margins:")
print(binary_margins)

print("\nGradient dW:")
print(dW)
```
